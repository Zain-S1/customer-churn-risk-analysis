#Data Wrangling

```python
#filling missing values
sales.fillna(5401, inplace=True)
```

```python
#checking for duplicates
sales.duplicated().sum()
```

```python
#fixing data types
sales['Order Date'] = pd.to_datetime(sales['Order Date'], format='%d/%m/%Y') #convert Order Date to datetime format

sales['Ship Date'] = pd.to_datetime(sales['Ship Date'], format='%d/%m/%Y') #convert Ship Dates to datetime format

sales['Postal Code'] = sales['Postal Code'].astype(str) #convert Postal Code to string type
sales['Postal Code'] = sales['Postal Code'].apply(lambda x: x.zfill(5) if len(x) < 5 else x) #putting back the leading zero
```

#Data Model Transformation

The `Star Schema` is the optimized structure for reporting and analysis in Power BI.

Therefore, this section transforms the single flat table into a **fact table** and several **dimension tables**.

###Order Dimension Table

```python
#order dimension table
order = sales[['Order ID', 'Order Date']]
order.drop_duplicates(inplace=True)

order.to_csv('Dim_Order.csv', index=False)
```

###Shipment Dimension Table

```python
#shipment dimension table
shipment = sales[['Ship Date', 'Ship Mode']].drop_duplicates().reset_index(drop=True)

#generate a ShipmentID as the ShipmentKey
shipment['ShipmentKey'] = shipment.index + 1 #start from 1

#reorder columns
shipment = shipment[['ShipmentKey', 'Ship Date', 'Ship Mode']]

shipment.to_csv('Dim_Shipment.csv', index=False)
```

###Customer Dimension Table

```python
#customer dimension table
customer = sales[['Customer ID', 'Customer Name', 'Segment']]
customer.drop_duplicates(inplace=True)

customer.to_csv('Dim_Customer.csv', index=False)
```

###Geography Dimension Table

```python
#geography dimension table
geography = sales[['Country', 'City', 'State', 'Postal Code', 'Region']].drop_duplicates().reset_index(drop=True)

#generate a GeoID as the GeographyKey
geography['GeographyKey'] = geography.index + 1 #start from 1

#reorder columns
geography = geography[['GeographyKey', 'Country', 'City', 'State', 'Postal Code', 'Region']]

geography.to_csv('Dim_Geography.csv', index=False)
```

###Product Dimension Table

```python
#product dimension table
product = sales[['Product ID', 'Category', 'Sub-Category', 'Product Name']]
product.drop_duplicates(inplace=True)

product.to_csv('Dim_Product.csv', index=False)
```

### Sales Fact Table

```python
#sales fact table
sales = sales[['Row ID', 'Order ID', 'Order Date', 'Ship Date', 'Customer ID', 'Product ID', 'Sales']]

sales.to_csv('Sales.csv', index=False)
```
